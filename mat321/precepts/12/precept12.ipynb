# Precept 12: Initial value problems
Today we will consider explicit and implicit methods. In particular,
we will consider two explicit methods: Euler's method
$$
y_{j+1} = y_j + h f(x_j,y_j)
$$
the improved Euler method:
$$
y_{j+1} = y_j + \frac{h}{2} \left(  f(x_j,y_j) +  f(x_{j+1},y_j + h f(x_j, y_j) \right)
$$

and two implicit methods: implicit Euler's
$$
y_{j+1} = y_j + h f(x_{j+1},y_{j+1}),
$$
and the trapezoidal rule method
$$
y_{j+1} = y_j + \frac{h}{2} \left(  f(x_{j},y_{j}) +  f(x_{j+1},y_{j+1}) \right)
$$
on two sets of equations.
